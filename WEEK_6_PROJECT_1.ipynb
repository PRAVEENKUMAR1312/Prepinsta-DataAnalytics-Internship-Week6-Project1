# **DATA CLEANING**

```python
featured_columns = ['Make','Model','Variant','Ex-Showroom_Price','Displacement','Cylinders','Valves_Per_Cylinder','Drivetrain','Emission_Norm','Fuel_Type','Seating_Capacity','Type','Wheelbase','Wheels_Size','Central_Locking','Number_of_Airbags','ABS_(Anti-lock_Braking_System)','EBD_(Electronic_Brake-force_Distribution)','ESP_(Electronic_Stability_Program)','Tyre_Pressure_Monitoring_System']
clean_data = pd.DataFrame(df[featured_columns])
clean_data
```

```python
# Converting 'Displacement' column to int
clean_data['Displacement'] = clean_data['Displacement'].str.replace('cc','')
clean_data['Displacement'].fillna(0, inplace=True)
clean_data['Displacement'] = clean_data['Displacement'].astype(int)
clean_data.loc[:,'Wheelbase'] = clean_data['Wheelbase'].str.replace(' mm','')
clean_data = pd.DataFrame(clean_data)
clean_data
```

```python
df['Make'] = df['Make'].replace('Land Rover Rover','Land Rover')
```

# **5.HISTOGRAM**

```python
numeric_columns = ['Ex-Showroom_Price','Displacement','Cylinders','Valves_Per_Cylinder','Seating_Capacity','Number_of_Airbags']
for col in numeric_columns:
    plt.hist(clean_data[col], color = 'green', edgecolor = 'black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(f'{col}')
    plt.ylabel('Frequency')
    plt.show()
```

# **6.BARCHARTS**


```python
top_10_car_names = clean_data['Car Name'].value_counts().head(10)
# Bar chart for the distribution of car makers (top 10)
plt.figure(figsize=(20,6))
plt.bar(top_10_car_names.index,top_10_car_names.values,color = 'blue')
plt.title('Top 10 Car Names Distribution')
plt.xlabel('Car Name')
plt.ylabel('Values')
plt.show()
```

# **7.BOX PLOTS**

```python
# Create box plots for each numeric variable
plt.figure(figsize=(15, 6))
for column in numeric_columns:
    plt.figure(figsize=(15, 6))
    #plt.subplot(num_rows, num_cols, i)
    sns.boxplot(x=clean_data[column], palette='muted')
    plt.title(f'Box Plot of {column}')
    plt.xlabel(f'{column}')
    plt.ylabel('Values')

plt.show()
```

# **8.PIE CHARTS**

```python
top_5_car_models = clean_data['Model'].value_counts().head(5)
explode = (0.1,0,0,0,0,)
plt.figure(figsize=(5,6))
# create a pie chart
plt.pie(top_5_car_models.values, explode=explode,labels = top_5_car_models.index,autopct = '%1.1f%%', startangle=140)
plt.title('Distribution of Top 5 Car Models')
plt.show()
```

```python
df['Fuel_Type'].value_counts().plot(kind='pie')
```

# **9.COUNTPLOT**

```python
categorical_columns = ['Drivetrain','Fuel_Type','Seating_Capacity','Type']
# Set up subplots based on the number of columns
fig, axes = plt.subplots(nrows=len(categorical_columns), ncols=1, figsize=(10, 5 * len(categorical_columns)))

# Create count plots for each selected column
for i, column in enumerate(categorical_columns):
    sns.countplot(x=clean_data[column], ax=axes[i], order = clean_data[column].value_counts().index)
    axes[i].set_title(f'Count plot of {column}')
    axes[i].set_xlabel(f'{column}')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

# **CORRELATION MATRIX**

```python
corr_matrix = clean_data[numeric_columns].corr()
# Visualize correlation matrix using heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,annot = True, cmap = 'viridis',fmt = '.2f', linewidth=0.5)
plt.title('Correlation matrix of numeric variables')
plt.show()
```

# **STACKED BAR CHARTS**

```python
top_7_model = clean_data['Car Name'].value_counts().nlargest(7).index

# Filter the DataFrame for the top makes
filtered_data = clean_data[clean_data['Car Name'].isin(top_7_model)]

# Create a stacked bar plot
plt.figure(figsize=(12, 8))
sns.countplot(x='Car Name', hue="Fuel_Type", data=filtered_data, palette="muted", order=top_7_model)
plt.title('Stacked Bar Plot for Top 7 car names and fuel type')
plt.show()
```

# **VIOLIN PLOTS**

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x="Type", y="Cylinders", data=clean_data, palette="colorblind")
plt.title('Violin Plot for Seating Capacity across Drivetrain')
plt.show()
```

# **POINT** **PLOTS**

```python
# Distribution of 'Type' and 'Seating Capacity' and their relationship using joint plots
sns.set(style="whitegrid")
sns.jointplot(x=clean_data['Type'], y=clean_data['Seating_Capacity'], data=clean_data, kind="scatter")
plt.show()
```
